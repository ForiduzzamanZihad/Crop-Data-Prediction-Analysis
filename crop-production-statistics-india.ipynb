# Creating Linear Regression Model

> Feature Selection & Creating dummy variables for object datatype within the dataset

```python
crop_data = pd.get_dummies(data=df)
```

```python
crop_data.head()
```

*Dividing data into Training and testing data*

```python
from sklearn.model_selection import train_test_split
```

```python
X = crop_data.drop('Production',axis=1)
X.head()
```

```python
y = crop_data['Production']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
X_test.shape
```

```python
y_test.shape
```

**Creating Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
crop_model = LinearRegression()
```

Training the Model

```python
crop_model.fit(X_train,y_train)
```

**Prediction**

```python
crop_predictions = crop_model.predict(X_test)
crop_predictions
```

```python
crop_model.coef_
```

```python
crop_model.intercept_
```

*Now eventhough we have a set of predictions with us we need a way to actually tell if the model is accuarte enough, let's test that using metrics*

```python
predicted_crop_val = pd.DataFrame({'Actual':y_test,'Predicted':crop_predictions})
predicted_crop_val
```

**We can see that from above table the predicted and actual values don't match**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df['Production'].mean()
```

```python
crop_predictions.mean()
```

```python
mean_absolute_error(y_test,crop_predictions)
```

```python
mean_squared_error(y_test,crop_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,crop_predictions))
```

```python
def mape(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs((actual - pred) / actual)) * 100

mape(y_test,crop_predictions)
```

> Checking the residual plots

```python
test_residuals = y_test - crop_predictions
```

```python
test_residuals
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='red')
plt.title('Residual plot of Testing data');
```

```python
r = r2_score(y_test,crop_predictions)
print("R2score to predict using Linear Regression is ",r)
```

Results of training data

```python
train_set_predictions = crop_model.predict(X_train)
train_set_predictions
```

```python
train_set_predictions.mean()
```

```python
sns.scatterplot(x=y_train,y=y_train-train_set_predictions)
plt.axhline(y=0,color='red')
plt.title('Residual plot of Training data');
```

From the metrics that we have used to test our model we can straight away say that Linear Regression was not a good choice of algo for this data set
The residual plot is a clear explaination of this, we can see the that the residual points are not randomly distributed around the actual points themself.
